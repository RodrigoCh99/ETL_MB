```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('sp.csv', sep=';')
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df['ESTADO'] = df.LINK.apply(lambda x: x.split('/')[4])
df['CIDADE'] = df.LINK.apply(lambda x: x.split('/')[5])
df['BAIRRO'] = df.LINK.apply(lambda x: x.split('/')[6])
df['BANCO'] = df.LINK.apply(lambda x: x.split('/')[7])
df['AGENCIA'] = df.LINK.apply(lambda x: x.split('/')[8])
```

```python
df_sl = df.drop(['ESTADO','LINK', 'NOME', 'BANCO', 'AGENCIA'], axis=1)
```

```python
gb = df_sl.groupby(['CIDADE']).count()
```

```python
gb = gb.T
```

```python
gb.to_csv('gb.csv')
```

```python
df_sp = df[df['CIDADE'] == 'sao-paulo']
```

```python
df_sp = df_sp.drop(['LINK', 'NOME', 'ESTADO', 'CIDADE', 'AGENCIA'], axis=1)
```

```python
df_sp.groupby(['BAIRRO']).count()
```
